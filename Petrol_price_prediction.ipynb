```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#Read .csv file
df=pd.read_csv('train_data.csv')
df.head(7)
```

```python
#data preprocessing
df.drop(['Date'],1,inplace=True)
```

```python
#output of data preprocessing
df.head(7)
```

```python
#creating a col 'Prediction'
prediction_days=30
df['Prediction']=df[['Petrol (USD)']].shift(-prediction_days)
df.head()
```

```python
#output with new created col
df.tail(7)
```

```python
x=np.array(df.drop(['Prediction'],1))
x=x[:len(df)-prediction_days]
nan_array=np.isnan(x) #part of data preprocessing removing NaN from the data x
not_nan_array= ~ nan_array
v=x[not_nan_array]
v.reshape(-1,1)
print(v)
```

```python
y=np.array(df['Prediction'])
nan_array=np.isnan(y)
not_nan_array= ~ nan_array #part of data preprocessing removing NaN from the data in y
o=y[not_nan_array]
o.reshape(-1,1)
print(o)
```

```python
#splitting the data as train and test 80% train 20% test
from sklearn.model_selection import train_test_split
x_train,x_test,d_train,d_test=train_test_split(v,o,test_size=0.2)
```

```python
#Hyperparameter tuning
from sklearn.svm import SVR
svr_rbf=SVR(kernel='rbf',C=50,gamma=20)
svr_rbf.fit(x_train.reshape(-1,1),d_train.reshape(-1,1))
```

```python
svm_prediction=svr_rbf.predict(x_test.reshape(-1,1))
print("Predicted value",svm_prediction.reshape(-1,1)) #predicted value 
print()
print("Actual value", d_test.reshape(-1,1)) #expected value
```

```python
#accuracy of train dataset
svr_rbf_confidence_train=svr_rbf.score(x_train.reshape(-1,1),d_train.reshape(-1,1))
print("svr_rbf_accuracy_train",svr_rbf_confidence_train)
```

```python
#accuracy of test dataset
svr_rbf_confidence_test=svr_rbf.score(x_test.reshape(-1,1),d_test.reshape(-1,1))
print("svr_rbf_accuracy_test",svr_rbf_confidence_test)
```

```python
#R^2 calculation
from sklearn.metrics import r2_score
print("R^2",float(r2_score(svm_prediction.reshape(-1,1),d_test.reshape(-1,1))))
```
